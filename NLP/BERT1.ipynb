```python
def remove_emoji(string):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)
```

### Carregando modelos do TensorFlow Hub

Usamos o universal-sentence-encoder-cmlm/multilingual-base, um codificador universal de frases que suporta mais de 100 idiomas. 
Ele é treinado usando um modelo de linguagem condicional mascarado, conforme descrito no documento de referência.

```python
handle_preprocess = "https://tfhub.dev/google/universal-sentence-encoder-cmlm/multilingual-preprocess/2"
handle_encoder = "https://tfhub.dev/google/universal-sentence-encoder-cmlm/multilingual-base/1"
```

### O modelo de pré-processamento
As entradas de texto precisam ser transformadas em ids de token numéricos e organizadas em vários tensores antes de serem inseridas no BERT. O TensorFlow Hub fornece um modelo de pré-processamento correspondente para cada um dos modelos BERT, que implementa essa transformação usando operações TF da biblioteca TF.text.

```python
bert_preprocess = hub.KerasLayer(handle_preprocess, name='preprocessing')
bert_encoder = hub.KerasLayer(handle_encoder, trainable=True, name='BERT_encoder')
```

### Definindo o modelo
Criando um modelo bem ajustado muito simples, com o modelo de pré-processamento, o modelo BERT selecionado, uma camada Densa e uma camada de Dropout.

```python
# Bert Layers
text_input = tf.keras.layers.Input(shape=(), dtype=tf.string, name='text')
preprocessed_text = bert_preprocess(text_input)
outputs = bert_encoder(preprocessed_text)

# Neural Network Layers
net = outputs['pooled_output']
net = tf.keras.layers.Dropout(0.1)(net)
net = tf.keras.layers.Dense(1, activation='sigmoid')(net)

# Final Model
bert_model = tf.keras.Model(text_input, net)
bert_model.summary()
```

### Treinamento de modelo
Agora temos todas as peças para treinar um modelo, incluindo o módulo de pré-processamento, codificador BERT, dados e classificador.

##### Função de Perda 
Quando o problema é de classificação binária e o modelo gera uma probabilidade (uma camada única unidade), você vai usar a função de perda losses.BinaryCrossentropy.

##### Otimizador
Para o ajuste fino, vamos usar o mesmo otimizador com o qual o BERT foi originalmente treinado: o "Adaptive Moments" (Adam). Este optimizador minimiza a perda de predição e faz regularização por decaimento de peso (sem uso de momentos), que também é conhecido como AdamW .


```python
loss = tf.keras.losses.BinaryCrossentropy(from_logits=True)
optimizer = tf.keras.optimizers.Adam(learning_rate=2e-5,epsilon=1e-08)

METRICS = [
    tf.keras.metrics.BinaryAccuracy(name='accuracy'),
    tf.keras.metrics.Precision(name='precision'),
    tf.keras.metrics.Recall(name='recall'),
]

bert_model.compile(loss=loss,optimizer=optimizer,metrics=[METRICS])
```

Quando o problema é de classificação binária e o modelo gera uma probabilidade (uma camada única unidade), você vai usar a função de perda losses.BinaryCrossentropy.

##### Otimizador
Para o ajuste fino, vamos usar o mesmo otimizador com o qual o BERT foi originalmente treinado: o "Adaptive Moments" (Adam). Este optimizador minimiza a perda de predição e faz regularização por decaimento de peso (sem uso de momentos), que também é conhecido como AdamW .


```python
loss = tf.keras.losses.BinaryCrossentropy(from_logits=True)
optimizer = tf.keras.optimizers.Adam(learning_rate=2e-5,epsilon=1e-08)

METRICS = [
    tf.keras.metrics.BinaryAccuracy(name='accuracy'),
    tf.keras.metrics.Precision(name='precision'),
    tf.keras.metrics.Recall(name='recall'),
]

bert_model.compile(loss=loss,optimizer=optimizer,metrics=[METRICS])
```
